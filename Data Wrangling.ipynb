```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Wrangling

```python
df = pd.read_csv("./dataset/SOF PUF 2023 - Clean.csv")
df.drop(columns=["Unnamed: 0", "Remittance Spent", "Past 6 Months Indicator"], inplace=True)
df = df.rename(columns={"Cash Home" : "Cash Brought Home"})
df.head()
```

```python
print("columns:", *df.columns, sep="\n")
```

```python
df['Marital Status'] = df['Marital Status'].replace('Unknown', np.nan)
```

```python
# Checking for missing values
df.isnull().sum()
```

```python
print(df["Education Level"].isnull().sum()/df.shape[0]*100)
# Education level is only missing 3.69% of its data, therefore the column still holds some signifance.
```

```python
# Imputing the missing values with the most frequent answer
df["Education Level"] = df["Education Level"].fillna(df["Education Level"].mode()[0])
df["Education Level"].unique()
```

```python
df["Usual Work"] = df["Usual Work"].replace(np.nan, "Unemployed")
df["Usual Work"].unique()
```

```python
print(df["Base"].isnull().sum()/df.shape[0]*100)
# The base is missing 16.93% of its data, still safe to be used.
df[df["Base"].isnull()]
# Base is NA due to tourist (Overseas Filipino Workers who simply traveled to be a tourist
```

```python
# changing base to N/A for tourist
df[df["Reason for Leaving"] == "Tourist"][["Reason for Leaving", "Base", "Worked Abroad", "Work"]]
# Since there are some tourist that found work abroad we cannot generalize the missing values are Not Applicable for base/work.
```

```python
is_tourist=df["Reason for Leaving"] == "Tourist"
tourist_who_worked= is_tourist & df["Industry"].notna()
true_tourist=is_tourist & df["Industry"].isna()
```

```python
df[tourist_who_worked][["Reason for Leaving", "Base", "Worked Abroad", "Work", "Industry", "Average Monthly Income", "Months Worked"]]
```

```python
df[true_tourist][["Reason for Leaving", "Base", "Work", 'Worked Abroad', "Industry", "Average Monthly Income", "Months Worked"]]
```

```python
df.loc[tourist_who_worked, "Worked Abroad"] = "Yes"
df.loc[tourist_who_worked & df.Base.isna(), "Base" ] = "Land-based"
df[tourist_who_worked][["Reason for Leaving", "Base", "Worked Abroad", "Work", "Industry", "Average Monthly Income", "Months Worked"]]
```

```python
df.loc[true_tourist, "Worked Abroad"] = "No"
df.loc[true_tourist, ["Base", "Industry", "Work"]] = "N/A"
df.loc[true_tourist, ["Average Monthly Income", "Months Worked"]] = 0.0
df[true_tourist][["Reason for Leaving", "Base", "Work", "Worked Abroad", "Industry", "Average Monthly Income", "Months Worked"]]
```

```python
df[~is_tourist & df["Base"].isna() & (df["Worked Abroad"] == "No")][["Reason for Leaving", "Base", "Work", "Worked Abroad", "Industry", "Average Monthly Income", "Months Worked"]]
```

```python
df.loc[~is_tourist & df["Base"].isna() & (df["Worked Abroad"] == "No"), ['Average Monthly Income', 'Months Worked']] = 0
df.loc[~is_tourist & df["Base"].isna() & (df["Worked Abroad"] == "No"), ['Work', 'Base', "Industry"]] = "N/A"
df[~is_tourist & (df["Base"] == "N/A") &  (df["Worked Abroad"] == "No")][["Reason for Leaving", "Base", "Work", "Worked Abroad", "Industry", "Average Monthly Income", "Months Worked"]]
```

```python
df[df["Base"].isna()][["Reason for Leaving", "Base", "Work", "Worked Abroad", "Industry", "Average Monthly Income", "Months Worked"]]
```

```python
df.loc[df["Base"].isna(), "Base"] = df["Base"].mode()[0]
df.Base
```

```python
df[df["Work"].isna()][["Reason for Leaving", "Base", "Worked Abroad", "Work", "Industry", "Average Monthly Income", "Months Worked"]]
```

```python
contradictory_rows = df['Reason for Leaving'].str.contains('OFW', na=False) & (df['Worked Abroad'] == 'No')
df = df[~contradictory_rows]
df[df["Work"].isna()][["Reason for Leaving", "Base", "Worked Abroad", "Work", "Industry", "Average Monthly Income", "Months Worked"]]
```

```python
df.isnull().sum()
```

```python
df[['Returned', "Month Returned", "Year Returned", "Return Reason", "Expected Month Returned", "Expected Year Returned"]]
```

```python
df[df['Returned']=="No"][['Returned', "Month Returned", "Year Returned", "Return Reason", "Expected Month Returned", "Expected Year Returned"]]
```

```python
not_returned = (df['Returned'] == "No") & (df["Month Returned"].isna()) & (df["Year Returned"].isna()) & (df['Return Reason'].isna())

df["Month Returned"].info()
df["Year Returned"].info()
df["Return Reason"].info()

df.loc[not_returned, ["Month Returned", "Return Reason"]] = np.nan
df.loc[not_returned, "Year Returned"] = np.nan

df[df['Returned']=="No"][['Returned', "Month Returned", "Year Returned", "Return Reason", "Expected Month Returned", "Expected Year Returned"]]
```

```python
df[(df['Returned']=="No") & (df["Reason for Leaving"] == "Immigrant")][["Reason for Leaving", "Expected Month Returned", "Expected Year Returned", "Return Reason"]]
```

```python
df.loc[~(df["Expected Month Returned"].isna()) & (df["Returned"]=="No") & (df["Reason for Leaving"]=="Immigrant") & ~(df["Expected Year Returned"].isna()) & (df["Return Reason"]=="N/A"), "Return Reason"] = "Unknown"

df.loc[(df["Expected Month Returned"].isna()) & (df["Returned"]=="No") & (df["Reason for Leaving"]=="Immigrant") & (df["Expected Year Returned"].isna()) & (df["Return Reason"]=="N/A"), "Expected Month Returned"] = "N/A"

df.loc[(df["Expected Month Returned"]=="N/A") & (df["Returned"]=="No") & (df["Reason for Leaving"]=="Immigrant") & (df["Expected Year Returned"].isna()) & (df["Return Reason"]=="N/A"), "Expected Year Returned"] = 0

df[(df['Returned'] == "No") & (df["Reason for Leaving"] == "Immigrant")][
    ["Reason for Leaving", "Expected Month Returned", "Expected Year Returned", "Return Reason"]]
```

```python
df[df["Expected Month Returned"].isna()][['Reason for Leaving','Returned', "Month Returned", "Year Returned", "Return Reason", "Expected Month Returned", "Expected Year Returned"]]
```

```python
already_returned = df['Returned'] == 'Yes'

df.loc[already_returned & df['Expected Month Returned'].isna(), 'Expected Month Returned'] = np.nan
df.loc[already_returned & df['Expected Year Returned'].isna(), 'Expected Year Returned'] = np.nan


still_abroad = df['Returned'] == 'No'

df.loc[still_abroad & df['Expected Month Returned'].isna(), 'Expected Month Returned'] = np.nan
df.loc[still_abroad & df['Expected Year Returned'].isna(), 'Expected Year Returned'] = np.nan

df[['Reason for Leaving','Returned', "Month Returned", "Year Returned", "Return Reason", "Expected Month Returned", "Expected Year Returned"]]
```
